# Graph Isomorphism Problem

The graph isomorphism problem is an important problem in graph theory and computer science. It is the problem of determining whether two given graphs are isomorphic or not. Two graphs are isomorphic if there exists a bijection (one-to-one and onto function) between their vertex sets that preserves the adjacency relation. In other words the graph isomorphism problem asks whether two graphs are structurally identical, meaning they have the same shape and connectivity pattern, even if the labels of the vertices and edges are different.

In this Jupyter notebook, we will explore the graph isomorphism problem and provide some examples using python.

### Graphs and determing their isopmophism

The networkx library in Python provides functions to determine if two given graphs are isomorphic.The time complexity of this algorithm is O ( n 2 log ‚Å° n + n m ), Here is an example code snippet:

```python
import networkx as nx
import matplotlib.pyplot as plt

G1 = nx.Graph([(0,1),(1,2),(2,0),(2,3)])
G2 = nx.Graph([(0,1),(1,2),(2,0),(2,3)])

# Draw Graph 1
nx.draw(G1, with_labels=True)
plt.show()

# Draw Graph 2
nx.draw(G2, with_labels=True)
plt.show()

if nx.is_isomorphic(G1, G2):
        print("Graphs are isomorphic")
else:
        print("Graphs are not isomorphic")
```

In graph theory, two graphs are isomorphic if they have the same number of vertices and edges, and their vertices and edges have the same connectivity pattern. In other words, two graphs are isomorphic if they can be transformed into each other by renaming their vertices.

The Python library NetworkX provides a function called is_isomorphic() that can be used to check if two graphs are isomorphic. In the given code, two example graphs G1 and G2 are created using the Graph() function of NetworkX. The first graph has vertices numbered from 0 to 3 and edges connecting vertices (0,1), (1,2), (2,0), and (2,3). The second graph is identical to the first one in terms of its connectivity pattern.

The is_isomorphic() function takes two graphs as input and returns a Boolean value indicating whether the two graphs are isomorphic or not. In the code, the is_isomorphic() function is used to check whether G1 and G2 are isomorphic. Since the two graphs have the same connectivity pattern and the same number of vertices and edges, the function returns True, indicating that the graphs are isomorphic.

If the two graphs were not isomorphic, the function would return False, indicating that the graphs have different connectivity patterns and are not equivalent under vertex renaming.

### Further examples

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create Graph 1
G1 = nx.Graph([(0,1),(0,3),(1,2),(1,4),(2,5),(3,4),(3,6),(4,5),(4,7),(5,8),(6,7),(7,8)])

# Create Graph 2
G2 = nx.Graph([(0,1),(0,6),(1,2),(1,7),(2,5),(3,4),(3,6),(4,5),(4,7),(5,8),(7,8)])

# Define a function to draw and compare the graphs
def draw_and_compare(G1, G2):
    # Draw Graph 1
    nx.draw(G1, with_labels=True)
    plt.show()

    # Draw Graph 2
    nx.draw(G2, with_labels=True)
    plt.show()

    # Check if graphs are isomorphic
    if nx.is_isomorphic(G1, G2):
        print("Graphs are isomorphic")
    else:
        print("Graphs are not isomorphic")

# Call the function to draw and compare the graphs
draw_and_compare(G1, G2)

# Modify Graph 2 to make it isomorphic to Graph 1
G2.remove_edge(0,6)
G2.add_edge(0,3)

# Call the function again to draw and compare the graphs
G2.clear()
G2.add_edges_from([(0,1),(0,3),(1,2),(1,7),(2,5),(3,4),(3,6),(4,5),(4,7),(5,8),(6,7),(7,8)])
draw_and_compare(G1, G2)

# Modify Graph 2 again to make it not isomorphic to Graph 1
G2.clear()
G2.add_edges_from([(0,1),(0,6),(1,2),(1,7),(2,5),(3,4),(3,6),(4,5),(4,7),(5,8),(7,8),(2,6)])
draw_and_compare(G1, G2)
```

## Bijections 

A bijection is a function that maps each element of a set to a unique element in another set.
Injective means we won't have two or more "A"s pointing to the same "B".

So many-to-one is NOT OK (which is OK for a general function).

As it is also a function one-to-many is not OK

But we can have a "B" without a matching "A"

Injective is also called "One-to-One"

Surjective means that every "B" has at least one matching "A" (maybe more than one).

There won't be a "B" left out.

Bijective means both Injective and Surjective together.

Think of it as a "perfect pairing" between the sets: every one has a partner and no one is left out.

So there is a perfect "one-to-one correspondence" between the members of the sets, In the context of the graph isomorphism problem, a bijection is used to map the vertices of one graph to the vertices of another graph in a way that preserves the adjacency relation.

# Preserving the adjecency relation

To determine whether two graphs are isomorphic, we need to find a bijection between their vertex sets that preserves the adjacency relation. The adjacency relation refers to the connection between two vertices in a graph.

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create two more complex graphs
G1 = nx.Graph()
G1.add_edges_from([(1, 2), (1, 3), (2, 4), (3, 4), (4, 5), (5, 6)])

G2 = nx.Graph()
G2.add_edges_from([('A', 'B'), ('A', 'C'), ('A', 'D'), ('B', 'C'), ('C', 'D'), ('D', 'E'), ('E', 'F')])

# Get adjacency matrices
A1 = nx.adjacency_matrix(G1).todense()
A2 = nx.adjacency_matrix(G2).todense()

# Plot the graphs and their adjacency matrices
fig, ax = plt.subplots(2, 2, figsize=(8, 8))

nx.draw(G1, ax=ax[0, 0], with_labels=True)
ax[0, 0].set_title('Graph 1')

nx.draw(G2, ax=ax[0, 1], with_labels=True)
ax[0, 1].set_title('Graph 2')

ax[1, 0].matshow(A1, cmap='binary')
ax[1, 0].set_title('Adjacency Matrix 1')

ax[1, 1].matshow(A2, cmap='binary')
ax[1, 1].set_title('Adjacency Matrix 2')

plt.show()
```

Based on the adjecency matrix above we can see that the graphs are not isomoprhic altough they have some similarites,  this just being another way to visualize the relationship between two graphs

### NP-intermediate

NP-intermediate (NPI) refers to a group of computational problems that belong to the complexity class NP, but are neither in class P or NP-complete, where P is considered easily solvable and NP hard to solve. The existence of these problems relies on the assumption that P and NP are not equal.

Ladner's theorem, proposed in 1975 states that if P and NP are not equal, then there must be NP-intermediate problems. The existence of NPI problems would also mean that P is not equal to NP. In other words, if there are no NP-intermediate problems P and NP would be equal.

Some problems, like the graph isomorphism problem, factoring and computing the discrete logarithm, are considered potential NP-intermediate problems.

Understanding the complexity of these problems can help us better design efficient algorithms and improve computational techniques.

Graph isomorphism is a good example of a potential NP-intermediate problem. It is an important concept in computer science, with real-world applications in various fields like chemistry.

In chemistry, for instance graph isomorphism is used to identify and compare molecular structures. Molecules can be represented as graphs, where atoms are the vertices and chemical bonds are the edges. The graph isomorphism problem arises when we need to determine whether two different molecules are, in fact, the same molecule with different labeling of atoms. If an efficient algorithm for solving graph isomorphism could be developed, it would help chemists quickly identify and classify molecules, aiding in drug discovery and research.

### Polynomial Time

Polynomial-time refers to a classification of algorithms based on their computational complexity, specifically the relationship between the input size and the number of steps or operations required to solve the problem. An algorithm is said to be polynomial-time if the maximum number of steps it takes to solve a problem is a polynomial function of the input size.

In other words, for an algorithm to be considered polynomial-time, its running time should grow at a rate proportional to a polynomial expression of the input size. A polynomial expression is typically represented as n^k, where n is the input size, k is a constant, and ^ denotes exponentiation.

Polynomial-time algorithms are considered efficient and practical for solving problems because their running time remains reasonable even as the input size increases. Conversely, algorithms with exponential or factorial time complexity tend to become impractical for larger input sizes due to the rapid growth in the number of required steps. In computer science, the class of problems solvable by polynomial-time algorithms is denoted by P.
