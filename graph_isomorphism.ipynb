# Graph Isomorphism Problem

### Graph Isomorphism Problem

The graph isomorphism problem is an important problem in graph theory and computer science. It is the problem of determining whether two given graphs are isomorphic or not. Two graphs are isomorphic if there exists a bijection (one-to-one and onto function) between their vertex sets that preserves the adjacency relation. In other words the graph isomorphism problem asks whether two graphs are structurally identical, meaning they have the same shape and connectivity pattern, even if the labels of the vertices and edges are different.

### Graphs and determing their isopmophism

The networkx library in Python provides functions to determine if two given graphs are isomorphic. Here is an example code snippet:

```python
import networkx as nx
import matplotlib.pyplot as plt

G1 = nx.Graph([(0,1),(1,2),(2,0),(2,3)])
G2 = nx.Graph([(0,1),(1,2),(2,0),(2,3)])

# Draw Graph 1
nx.draw(G1, with_labels=True)
plt.show()

# Draw Graph 2
nx.draw(G2, with_labels=True)
plt.show()
```

In graph theory, two graphs are isomorphic if they have the same number of vertices and edges, and their vertices and edges have the same connectivity pattern. In other words, two graphs are isomorphic if they can be transformed into each other by renaming their vertices.

The Python library NetworkX provides a function called is_isomorphic() that can be used to check if two graphs are isomorphic. In the given code, two example graphs G1 and G2 are created using the Graph() function of NetworkX. The first graph has vertices numbered from 0 to 3 and edges connecting vertices (0,1), (1,2), (2,0), and (2,3). The second graph is identical to the first one in terms of its connectivity pattern.

The is_isomorphic() function takes two graphs as input and returns a Boolean value indicating whether the two graphs are isomorphic or not. In the code, the is_isomorphic() function is used to check whether G1 and G2 are isomorphic. Since the two graphs have the same connectivity pattern and the same number of vertices and edges, the function returns True, indicating that the graphs are isomorphic.

If the two graphs were not isomorphic, the function would return False, indicating that the graphs have different connectivity patterns and are not equivalent under vertex renaming.

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create Graph 1
G1 = nx.Graph([(0,1),(0,3),(1,2),(1,4),(2,5),(3,4),(3,6),(4,5),(4,7),(5,8),(6,7),(7,8)])

# Create Graph 2
G2 = nx.Graph([(0,1),(0,6),(1,2),(1,7),(2,5),(3,4),(3,6),(4,5),(4,7),(5,8),(7,8)])

# Check if graphs are isomorphic
if nx.is_isomorphic(G1, G2):
    print("Graphs are isomorphic")
else:
    print("Graphs are not isomorphic")

# Draw Graph 1
nx.draw(G1, with_labels=True)
plt.show()

# Draw Graph 2
nx.draw(G2, with_labels=True)
plt.show()

```
